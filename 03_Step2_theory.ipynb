- If $\Delta x$ is small, it approximates $\frac{\partial u}{\partial x}$ well. The smaller $\Delta x$, the better the approximation
- Errors are always introduced (truncation errors)
- $\textbf{Order of accuracy (of difference approximation):}$ The power of $\Delta x$ with which truncation error -> 0
- Using Taylor Series Expansion, saw that Forward Difference (FD) and Backward Difference (BD) approximations both are 1st order or, according to "big-O notation" : $\mathbf{O}(\Delta x)$ 

## Difference Formula for first derivatives

- FD and BD are both called "one-sided" formulas because they involve points on only one-side of the point i that you try to evaluate from.

- The power of $\Delta x$ is one less than the power of the Taylor Expansion term as we divide with $\Delta x$.
- FD vs BD has difference in sign 
- We note that the dominant first order truncation terms are of opposite signs. We can add FD and BD to find the second order finite difference approximation. 
- FD + BD = 2$\frac{\partial u}{\partial x}$ which means that R.H.S of FD + BD is divided by 2. Therefore:

- Image above is central difference approximation. Truncation error of $\mathbf{O}(\Delta x^2)$
> Note: Consider a 1D domain b/w (0,1), with 11 mesh points, so $\Delta x = 0.1$. 11 points as starts from 0 and ends at 1. 1$^{st}$ order ~ $O(\Delta x)$ ~ $O$(10%). 2$^{nd}$ order ~ $O(\Delta x^2)$ ~ $O$(1%)

- For 1% accuracy with first order, would need 100 divisions, or 101 mesh points. $\Delta x$ = 0.01
- So higher cost with 1st order approximation to obtain same type of accuracy. Have to apply the difference formulation 100 times instead of 10 times, so lot more computation work.
- Interesting to note: First finite difference formula for $\frac{\partial u}{\partial x}|_i$ is actually the central difference for the midpoint.
- $\frac{\partial u}{\partial x}|_{i+\frac{1}{2}} = \frac{{u_{i+1} - u_{i}}}{\Delta x} + O(\Delta x^2)$
- $\frac{\partial u}{\partial x}|_{i-\frac{1}{2}} = \frac{{u_{i} - u_{i-1}}}{\Delta x} + O(\Delta x^2)$
- Gained an order of accuracy with the SAME expression as before


## Simplified Model Equation 

### Navier-Stokes Equation

- $\frac{\partial \overrightarrow{V}}{\partial t} + (\overrightarrow{V} \centerdot \bigtriangledown)\overrightarrow{V} = -\frac{\bigtriangledown p}{\rho} + \nu \bigtriangledown ^{2} \overrightarrow{V}$

- Diffusive terms, such as the last term, appear through second order derivatives
- Convective fluxes appear as first order derivatives in space
- Time variable also appears as first order derivative
- With this, many modelling assumptions can be made, such as steady flow making the first term 0, or ideal flow so no viscosity etc
    - This leads to various simplified model equations, but they are all a system of P.D.Es
    - The highest space derivative is second order and highest time derivative is first order.
- Have to develop numerical methods which allow us to analyse the mathematical properties of P.D.Es in different forms
- How do we recognise if convective phenomena, or diffusive phenomena? 
- What are the other phenomena that could occur?
- What are the associated boundary and initial conditions?
- Numerical discretisation must be adequate for the physical process $\rightarrow$ convection, diffusion, unsteadiness, so we benefit by studying model equations, which are simplified forms of the basic convection, diffusion equations
